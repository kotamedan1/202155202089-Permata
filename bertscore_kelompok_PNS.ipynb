```python
import torch
from transformers import BertTokenizer, BertForSequenceClassification
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import numpy as np
```

```python
model_name = 'bert-base-uncased'
tokenizer = BertTokenizer.from_pretrained(model_name)
model = BertForSequenceClassification.from_pretrained(model_name)
```

```python
texts = ["I love programming.", "Python is great.", "Machine learning is fascinating.",
         "I enjoy learning new things.", "Artificial intelligence is the future."] * 20  # Repeated to have more samples
labels = [1, 1, 1, 1, 1] * 20  # Dummy labels
```

```python
inputs = tokenizer(texts, return_tensors='pt', padding=True, truncation=True)
input_ids = inputs['input_ids']
attention_mask = inputs['attention_mask']
```

```python
labels = torch.tensor(labels)
```

```python
def calculate_metrics(true_labels, pred_labels):
    accuracy = accuracy_score(true_labels, pred_labels)
    precision = precision_score(true_labels, pred_labels, average='binary')
    recall = recall_score(true_labels, pred_labels, average='binary')
    f1 = f1_score(true_labels, pred_labels, average='binary')
    return accuracy, precision, recall, f1
```

```python
num_tests = 10
accuracy_scores = []
precision_scores = []
recall_scores = []
f1_scores = []
```

```python
for i in range(num_tests):
    train_idx, test_idx = train_test_split(range(len(texts)), test_size=0.2, random_state=i)
    # The following lines were indented too far, causing the error
    train_inputs = input_ids[train_idx]
    train_labels = labels[train_idx]
    train_mask = attention_mask[train_idx]

    test_inputs = input_ids[test_idx]
    test_labels = labels[test_idx]
    test_mask = attention_mask[test_idx]
```

```python
model.eval()
with torch.no_grad(): # Removed extra indentation here
    outputs = model(input_ids=test_inputs, attention_mask=test_mask)
    logits = outputs.logits
    predictions = torch.argmax(logits, dim=-1)

accuracy, precision, recall, f1 = calculate_metrics(test_labels, predictions)
accuracy_scores.append(accuracy)
precision_scores.append(precision)
recall_scores.append(recall)
f1_scores.append(f1)
```

```python
avg_accuracy = np.mean(accuracy_scores)
avg_precision = np.mean(precision_scores)
avg_recall = np.mean(recall_scores)
avg_f1 = np.mean(f1_scores)

print(f'Average Accuracy: {avg_accuracy}')
print(f'Average Precision: {avg_precision}')
print(f'Average Recall: {avg_recall}')
print(f'Average F1 Score: {avg_f1}')
```
